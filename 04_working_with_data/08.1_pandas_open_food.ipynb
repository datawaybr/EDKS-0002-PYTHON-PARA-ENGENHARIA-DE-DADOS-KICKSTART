```python
import pandas as pd
import requests
from datetime import datetime as dt
pd.set_option('display.max_colwidth', None)
```

```python
url = "https://world.openfoodfacts.net/api/v2/product/"
product = '7891000369371'
```

```python
url_completa = url + product
```

```python
url_completa
```

```python
resposta = requests.get(url_completa)
```

```python
resposta.status_code
```

```python
crunch = resposta.json()
```

```python
crunch
```

```python
df = pd.json_normalize( crunch.get('product') )
```

```python
df
```

```python
df_cols = df[['code','product_name','image_front_url', 'last_modified_t']].copy()
```

```python
df_cols
```
